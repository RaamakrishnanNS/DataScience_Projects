#### Observations:
1. For this analysis Umpire 3 is not required. We can remove the column using drop()
2. Out of 756 matches, 4 matches ended with no result. This could be missing data or match stopped due to rain.

```python
# Dropping Umpire3 column
matches_df = matches_df.drop('umpire3', axis=1)
matches_df.head()
```

#### Number of matches and teams

1. Trying to find number of matches played in each season in the IPL from its inception to 2019

```python
# We are using Groupby function on 'Season' column and counting number of id. You can also select any other column name
matches_per_season = matches_df.groupby('season').id.count()

matches_per_season
```

```python
# We are using Groupby function on 'Season' column and counting number of umpire2. You can also select any other column name
matches_per_season_umpire2 = matches_df.groupby('season').umpire2.count()

matches_per_season_umpire2
```

##### Plotting the results

```python
# Plotting using barplot() method
# Setting the figure parameters like figsize, xticks(), rotation etc
plt.figure(figsize=(12,6))

# This command will turn the x labels by 75 degrees to make it more readable
plt.xticks(rotation=75)
plt.title('Matches Per Season')
match_per_season_plot = sns.barplot(x= matches_per_season.index, y=matches_per_season)
match_per_season_plot.set(xlabel = 'Season', ylabel='No. of Matches')
```

#### Analyzing the Toss results

1. Analyzing the Toss decision column using value_counts()

```python
toss_decision = matches_df.groupby('season').toss_decision.value_counts().sort_index()/matches_per_season *100
```

```python
toss_decision.index
```

```python
# To use toss decision and season both as an index, use unstack() option
toss_decision.unstack()
```

```python
# Plotting toss decisions
#toss_decision.unstack().plot(kind = 'bar', figsize=(12,6), title = 'Toss Decisions', xlabel = 'Seasons', ylabel = 'Percentage');
toss = sns.barplot(x= toss_decision.index, y=toss_decision)
toss.set(xlabel = 'Season', ylabel='No. of Matches')
```

```python
toss_decision.unstack().plot(kind = 'bar', figsize=(12,6), title = 'Toss Decisions')
```

### Number of Wins

```python
# Filtering wins based on First batting
win_by_batting_first = matches_df[(matches_df.win_by_wickets == 0) & (matches_df.result == 'normal')].groupby('season').winner.count()/matches_per_season*100
```

```python
# Filtering wins based on Second batting
win_by_batting_second = matches_df[(matches_df.win_by_runs == 0) & (matches_df.result == 'normal')].groupby('season').winner.count()/matches_per_season*100
```

```python
# Season wise information on % wins batting first and % wins batting second
season_wise_info = pd.concat([win_by_batting_first, win_by_batting_second], axis=1)
season_wise_info.columns = ['Batting First', 'Batting Second']
season_wise_info
```

```python
# Plotting the results
season_wise_info.plot(kind='bar', figsize=(12,6), title='Wins')
```

### Teams with "History"

```python
# Total matches played by each team
total_matches_played = (matches_df.team2.value_counts() + matches_df.team1.value_counts()).sort_values(ascending = False)
total_matches_played
```

```python
# Plotting the data
plt.figure(figsize=(12,6))
plt.title('Total Matches Played')
total_matches_played_plot = sns.barplot(y = total_matches_played.index, x = total_matches_played)
total_matches_played_plot.set(ylabel = 'Teams', xlabel = 'No. of Matches');
```

### Teams with "Legacy"

```python
# Teams with Most Wins
most_wins = matches_df.winner.value_counts()

most_wins
```

```python
win_percentage = (most_wins / total_matches_played).sort_values(ascending = False) * 100
win_percentage
```

```python
plt.figure(figsize=(12,6))
plt.title('Win Percentage')
win_percentage_plot = sns.barplot(y = win_percentage.index, x = win_percentage)
total_matches_played_plot.set(ylabel = 'Teams', xlabel = 'Percentage');
```

### Questions:

#### Who has won the IPL tournament?

```python
ipl_win = matches_df.groupby('season').tail(1).sort_values('season', ascending = True)
ipl_win
```

```python
ipl_winners = ipl_win.winner.value_counts()
ipl_winners
```

```python
plt.figure(figsize=(18, 4))
plt.xlabel('Teams')
plt.ylabel('No. of Times')
plt.title('IPL Champions')
sns.barplot( x  = ipl_winners.index, y = ipl_winners);
```

#### Which are the most and least consistent teams across all the seasons?

```python
matches_won_each_season = pd.crosstab(matches_df['winner'], matches_df['season'])

matches_won_each_season
```

```python
plt.figure(figsize=(15, 9))
plt.xlabel('Seasons')
plt.ylabel('Teams')
plt.title('Matches Won Each Season')
sns.heatmap(matches_won_each_season, annot = True)
```

#### What has been the biggest margin of victory in terms of runs in the IPL?

```python
highest_wins_by_runs_df = matches_df[matches_df.win_by_runs != 0].sort_values('win_by_runs', ascending = False)
highest_wins_by_runs_df
```

```python
plt.figure(figsize=(25, 15))
plt.xlabel('Seasons',size=30)
plt.ylabel('Runs',size=30)
plt.title('Highest Wins By Runs', size = 30)
sns.scatterplot(x = 'season',y =  'win_by_runs', data = highest_wins_by_runs_df, s =150, color = 'black');
sns.scatterplot(x = 'season',y =  'win_by_runs', data = highest_wins_by_runs_df.head(10), s =220, color = 'red');
for i in range(highest_wins_by_runs_df.head(10).shape[0]):
    plt.annotate(highest_wins_by_runs_df.winner.tolist()[i], (highest_wins_by_runs_df.season.tolist()[i]+0.1, highest_wins_by_runs_df.win_by_runs.tolist()[i] - 1) , size = 20)
```
